```python
x=d_train_sample.drop(columns="label")
x
```

```python
y=d_train_sample["label"]
y
```

```python
for i in range(5):
    plt.subplot(5,1,i+1)
    plt.imshow(x.iloc[i].values.reshape(28,28),cmap='gray')
    plt.show()
```

```python
plt.figure(figsize=(28, 23))
for i in range(5):
    plt.subplot(5, 1, i+1)
    plt.imshow(x.iloc[i].values.reshape(28,28), cmap='gray')
    plt.title(f"Label: {y.iloc[i]}")
    plt.axis('off')
#plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_val,y_train,y_val=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train.shape,x_val.shape,y_train.shape,y_val.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3,metric="minkowski",p=2)
#knn=KNeighborsClassifier(n_neighbors=3,metric="euclidean")
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_val)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

cm=confusion_matrix(y_val,y_pred)
print("Confusion matrix:\n",cm)
print("\n")
acc=accuracy_score(y_val,y_pred)
print("Accuracy:",acc)
print("\n")
cr=classification_report(y_val,y_pred)
print("classification report:\n",cr)
```

```python
cm = confusion_matrix(y_val, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
d_test = pd.read_csv("Week7_test.csv")
X_test = d_test
#y_test = d_test['label']

y_test_pred = knn.predict(X_test)

plt.figure(figsize=(10, 3))
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(X_test.iloc[i].values.reshape(28, 28), cmap='gray')  # Adjust shape if needed
    plt.title(f"Pred: {y_test_pred[i]}")
    plt.axis('off')
plt.tight_layout()
plt.show()
```
